# McNemar's test

```python
# importerer mcnemar
from statsmodels.stats.contingency_tables import mcnemar
```

### Overall


```python
models = ['GPT-4o', 'GPT-4', 'CoPilot']

# laver en liste over alle mulige par
pairs = [(models[i], models[j]) for i in range(len(models)) for j in range(i + 1, len(models))]

# Laver McNemar's test for hver eneste par i "overall" data
for model1, model2 in pairs:
    print(f"Comparing {model1} and {model2}")
    contingency_table = pd.crosstab(data[model1], data[model2])
    print(contingency_table)
    result = mcnemar(contingency_table, exact=True)
    print(f"Statistic: {result.statistic}, p-value: {result.pvalue}\n")
```

### General McNemar

```python
import pandas as pd
import statsmodels.api as sm
from statsmodels.stats.contingency_tables import mcnemar



# Laver kombinationer af  is_MC og is_visual
combinations = [
    (1, 1),  # is_MC og is_visual
    (1, 0),  # is_MC og not is_visual
    (0, 1),  # not is_MC og is_visual
    (0, 0)   # not is_MC og not is_visual
]

# laver en liste over alle mulige par
models = ['GPT-4o', 'GPT-4', 'CoPilot']
pairs = [(models[i], models[j]) for i in range(len(models)) for j in range(i + 1, len(models))]

# Laver McNemar's test for hver eneste par for alle underkategorier
for is_mc, is_visual in combinations:
    subset = data[(data['is_MC'] == is_mc) & (data['is_visual'] == is_visual)]
    print(f"\nSubset: is_MC={is_mc}, is_visual={is_visual}")
    
    for model1, model2 in pairs:
        if len(subset) > 0:
            print(f"  Comparing {model1} and {model2}")
            contingency_table = pd.crosstab(subset[model1], subset[model2])
            print(contingency_table)  # printer contingency tables til tabel senere
            result = mcnemar(contingency_table, exact=True)
            print(f"    Statistic: {result.statistic}, p-value: {result.pvalue:.3f}")
        else:
            print(f"  No data for comparison between {model1} and {model2} in this subset")
```

# Binomialfordeling


```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import binom

# henter data
data = pd.read_csv('GMperformance.csv', delimiter=';')
data = data[data['Year'] != 2007]

# vælger alle rækker hvor det er multiple choice
mc_data = data[data['is_MC'] == 1]

# Laver tupler til opbevaring af acc og CI
models = ['GPT-4o', 'GPT-4', 'CoPilot']
accuracies = {}
confidence_intervals = {}

n_trials = mc_data.shape[0]

for model in models:
    accuracy = mc_data[model].mean()
    accuracies[model] = accuracy
    
    # beregner standard error
    se = np.sqrt(accuracy * (1 - accuracy) / n_trials)
    
    # beregner 95%-konfidensinterval, bruger "normal approximation"
    ci_lower = accuracy - 1.96 * se
    ci_upper = accuracy + 1.96 * se
    confidence_intervals[model] = (ci_lower, ci_upper)

# beregner baseline accuracy og konfidensinterval
baseline_p = 0.2  # pga 1/5 er tilfældigt gæt
baseline_se = np.sqrt(baseline_p * (1 - baseline_p) / n_trials)
baseline_ci_lower = baseline_p - 1.96 * baseline_se
baseline_ci_upper = baseline_p + 1.96 * baseline_se
confidence_intervals['Baseline'] = (baseline_ci_lower, baseline_ci_upper)

# Plotter binomialfordelinger
plt.figure(figsize=(12, 8))

x = np.arange(0, n_trials + 1)
colors = ['b', 'g', 'r']

for idx, model in enumerate(models):
    p = accuracies[model]
    binom_dist = binom.pmf(x, n_trials, p)
    plt.plot(x, binom_dist, label=f'{model} (p={p:.2f})', color=colors[idx])
    
# Plotter baseline performance (tilfældigt gæt)
baseline_dist = binom.pmf(x, n_trials, baseline_p)
plt.plot(x, baseline_dist, 'k--', label='Baseline (p=0.2)')

# Plotter de tre modellers CI som horisontale barer
offsets = np.linspace(0.05, 0.07, len(models) + 1)
for idx, (model, color, offset) in enumerate(zip(models, colors, offsets)):
    ci_lower, ci_upper = confidence_intervals[model]
    y = 0.015 + offset  # Rykker CI op en smule
    plt.hlines(y=y, xmin=int(n_trials * ci_lower), xmax=int(n_trials * ci_upper), color=color, linewidth=2)
    plt.plot([int(n_trials * ci_lower), int(n_trials * ci_upper)], [y, y], 's', color=color, markersize=5)
    plt.plot(int(n_trials * accuracies[model]), y, 'o', color=color, markersize=5)
    plt.scatter([], [], color=color, marker='s', label=f'{model} CI') 


# Plotter baseline CI
ci_lower, ci_upper = confidence_intervals['Baseline']
y = 0.011 + offsets[-1]
plt.hlines(y=y, xmin=int(n_trials * ci_lower), xmax=int(n_trials * ci_upper), color='gray', linewidth=2)
plt.plot([int(n_trials * ci_lower), int(n_trials * ci_upper)], [y, y], 's', color='gray', markersize=5)
plt.plot(int(n_trials * baseline_p), y, 'o', color='gray', markersize=5)
plt.scatter([], [], color='gray', marker='s', label='Baseline CI')


# Indstillinger
plt.xlim([25, 105])
plt.ylim([0, 0.09])
plt.xlabel(f'Number of Successful Trials Out of {n_trials} MC Questions')
plt.ylabel('Probability')
plt.title('Binomial Distributions of Model Accuracies (MC = True)')
plt.legend(loc='lower right')
plt.grid(True)
plt.show()
```
